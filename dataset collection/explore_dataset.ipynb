```python
genius = Genius(GENIUS_API_KEY,
                skip_non_songs=True,
                remove_section_headers=True,
                verbose=False,
                sleep_time=1,
                retries=3
            )

def get_song(name, artist=None):
    try:
        # print(f"Doing {name} by {artist}", end='\r', flush=True)
        pattern = r'^[\s\S]*?Lyrics\s*|(\d*)Embed$|See.*Live|Get tickets.*?\$[\d,]+|You might also like'
        song = genius.search_song(name, artist) if artist else genius.search_song(name)
        return re.sub(pattern, '', song.lyrics, flags=re.MULTILINE)
    except TimeoutError:
        time.sleep(10)
        return get_song(name, artist)
```

```python
empty = {
    'danceability': None,
    'energy': None,
    'key': None,
    'loudness': None,
    'mode': None,
    'speechiness': None,
    'acousticness': None,
    'instrumentalness': None,
    'liveness': None,
    'valence': None,
    'tempo': None,
    'duration_ms': None,
    'time_signature': None
}

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=SPOTIFY_CLIENT_ID, 
                                                           client_secret=SPOTIFY_CLIENT_SECRET),
                    requests_timeout=15, retries=5)

def get_audio_features(song_name, artist_name):
    # print(f"Doing {song_name} by {artist_name}", end='\r', flush=True)
    
    results = sp.search(q=f'track:{song_name} artist:{artist_name}', type='track', limit=1)
    if results['tracks']['items']:
        track = results['tracks']['items'][0]
        track_id = track['id']
        
        audio_features = sp.audio_features(track_id)
        
        if audio_features:
            audio_features = audio_features[0]
            audio_features.pop('analysis_url')
            audio_features.pop('track_href')
            audio_features.pop('id')
            audio_features.pop('uri')
            audio_features.pop('type')
            return audio_features
        
    return empty
```

Fixing Lyrics

```python
import os
import pandas as pd
from fuzzywuzzy import fuzz, process

# Path to the directory containing text files
directory = "manually_check"

# Function to extract song and artist from filename
def get_song_artist(filename):
    name = os.path.splitext(filename)[0]  # Remove extension
    if '-' in name:
        song, artist = name.split('-', 1)  # Split at the first dash
        return song.strip(), artist.strip()
    return None, None

# Read lyrics text from file
def read_lyrics(filepath):
    with open(filepath, 'r', encoding='utf-8') as file:
        return file.read()

# Process files and update lyrics column in DataFrame
for filename in os.listdir(directory):
    song, artist = get_song_artist(filename)
    if song and artist:
        file_path = os.path.join(directory, filename)
        lyrics = read_lyrics(file_path)
        
        # Fuzzy matching for song and artist in the dataset
        match_song, _ , _ = process.extractOne(song, df['name'], scorer=fuzz.token_sort_ratio)
        match_artist, _ , _ = process.extractOne(artist, df['artist'], scorer=fuzz.token_sort_ratio)
        
        # Find the row in DataFrame matching the song and artist
        match_row = df[(df['name'] == match_song) & (df['artist'] == match_artist)]
        
        if not match_row.empty:
            # Update the lyrics column for the matching row
            df.loc[match_row.index, 'lyrics'] = lyrics
        else:
            print(song, artist)
```
